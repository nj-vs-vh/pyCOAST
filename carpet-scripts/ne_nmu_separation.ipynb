```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.size": 16, 
    "font.sans-serif": ["Helvetica"],
})
```

```python
photons = pd.read_csv("./data/photons-ne-vs-nmu.txt", sep="\t")
protons = pd.read_csv("./data/protons-ne-vs-nmu.txt", sep="\t")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.scatter(photons['N_e'], photons['N_mu'], color='red', label='$\gamma$ primary')
ax.scatter(protons['N_e'], protons['N_mu'], color='blue', label='p primary')

ax.set_xlabel('$N_{e^{+/-}}$')
ax.set_ylabel('$N_{\mu^{+/-}}$ ($E > 1$ GeV, $R_{axis} < 100$ m)')
ax.legend()

# ax.set_yscale('log')
# ax.set_xscale('log')
plt.show()
```
